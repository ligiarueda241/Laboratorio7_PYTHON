```python
# Realizar un programa que recorra la lista de números
for num in range(1, 6):  # Números del 1 al 5
    if num == 4:
        print("Número 4 encontrado, deteniendo el bucle.")
        break  # Detener el bucle cuando num es 4

if num % 2 == 0:
    print(f"{num} es par")
else:
    print(f"{num} es impar")

i = 0
while i < 3:
    print(f"Contando: {i}")
    i += 1
```
